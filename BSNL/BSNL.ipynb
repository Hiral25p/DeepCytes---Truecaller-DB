```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import matplotlib.font_manager as fm
```

```python
plt.rcParams['font.family'] = 'DejaVu Sans'
```

```python
file_path = '19M-BSNL_Mobile_sample.csv'

# Specify dtype and set low_memory to False
dtype_spec = {0: 'str', 4: 'str', 8: 'str', 9: 'str'}

df = pd.read_csv(file_path, dtype=dtype_spec, low_memory=False, on_bad_lines='skip')
```

```python
sns.set(style="whitegrid")
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
missing_values_percentage = (missing_values / len(df)) * 100
print(missing_values_percentage)
```

```python
# Visualize missing values using a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Distribution of 'Carrier'
plt.figure(figsize=(10, 6))
sns.countplot(y='Carrier', data=df, order=df['Carrier'].value_counts().index)
plt.title('Distribution of Carrier')
plt.show()
```

```python
# Distribution of 'Address'
plt.figure(figsize=(10, 6))
sns.countplot(y='Address', data=df, order=df['Address'].value_counts().index)
plt.title('Distribution of Address')
plt.show()
```

```python
# Distribution of 'JobTitle'
plt.figure(figsize=(10, 6))
sns.countplot(y='JobTitle', data=df, order=df['JobTitle'].value_counts().index)
plt.title('Distribution of JobTitle')
plt.show()
```
